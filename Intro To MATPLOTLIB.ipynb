```python
## WORKFLOW
#import matplot lib and get it ready for plotting
%matplotlib inline
import matplotlib.pyplot as plt

#prepare the data
x=[1,2,3,4]
y=[10,20,30,40]

#setup that data
fig, ax = plt.subplots(figsize=(10,10))

#plot data
ax.plot(x,y)

#customize plot
ax.set(title="Simple Plot",
       xlabel="x-axis",
      ylabel="y-axis")
```

```python
import numpy as np
#create some data
x = np.linspace(0,5,10)
x
```

```python
# data representation using line plot
fig, ax = plt.subplots()
ax.plot(x,  x**2)
```

```python
# data representation using scatter plot
fig, ax = plt.subplots()
ax.scatter(x,x**2);
```

```python
fig, ax = plt.subplots()
ax.scatter(x,np.sin(x));
```

```python
#plotting a dictonary using a bar plot
chocolates_rating = { "5 star":20,
                    "dairymilk": 15,
                    "kitkat":30}

fig, ax = plt.subplots()
ax.bar(chocolates_rating.keys(),chocolates_rating.values());

ax.set(title= "Chocolate Ratings",
      xlabel= "Chocolates",
      ylabel= "Ratings");
```

```python
# for horizontal bar you have to put (x,y) in list to show the o/p
fig, ax = plt.subplots()
ax.barh(list(chocolates_rating.keys()),list(chocolates_rating.values()))
ax.set(title= "Chocolate Ratings",
      ylabel= "Chocolates",
      xlabel= "Ratings");
```

```python
# random data with histogram
x = np.random.randn(1000)
fig, ax = plt.subplots()
ax.hist(x);
```
