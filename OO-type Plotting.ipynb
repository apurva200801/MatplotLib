```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
heart_disease = pd.read_csv("heart-disease.csv")



```

```python
heart_disease.head()
```

```python
#Ploting the OO type Format:
fig, ax = plt.subplots(figsize=(10,6))

#plotting the data:
scatter = ax.scatter(x=heart_disease["age"],
                    y=heart_disease["chol"],
                    c=heart_disease["target"])

#customizing the plot:
ax.set(title="Heart disease and chol level",
       xlabel="Age",
      ylabel="Chol")

#adding a legend
ax.legend(*scatter.legend_elements(),title="target");

#adding a horizontal line:
ax.axhline(heart_disease["chol"].mean(),
          linestyle='--');
```

```python

plt.style.use("seaborn")

# plotting subplots with OO:
#Ploting the OO type Format:
fig, (ax0,ax1) = plt.subplots(nrows=2,
                              ncols=1,
                              figsize=(10,10))
#Adding data to ax0:
#plotting the data:
scatter = ax0.scatter(x=heart_disease["age"],
                    y=heart_disease["chol"],
                    c=heart_disease["target"],
                     cmap="winter")

#customizing the plot:
ax0.set(title="Heart disease and chol level",
       xlabel="Age",
      ylabel="Chol")

#adding a legend
ax0.legend(*scatter.legend_elements(),title="target");

#adding a horizontal line:
ax0.axhline(heart_disease["chol"].mean(),
          linestyle='--');


#adding data to ax1:
#plotting the data:
scatter = ax1.scatter(x=heart_disease["age"],
                    y=heart_disease["thalach"],
                    c=heart_disease["target"],
                     cmap="winter")

#customizing the plot:
ax1.set(title="Heart disease and thalach level",
       xlabel="Age",
      ylabel="Thalach")

#adding a legend
ax1.legend(*scatter.legend_elements(),title="target");

#adding a horizontal line:
ax1.axhline(heart_disease["thalach"].mean(),
          linestyle='--');

```
